```
import tensorflow as tf

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/tmp/data/", one_hot=True) # mnist 데이터 다운로드

x = tf.placeholder(tf.float32 , shape = [None , 784]) # 입력값을 받기 위해서 설정
y = tf.placeholder(tf.float32, shape = [None , 10]) # 출력 값을 받기 위해서 설정 

W = tf.Variable(tf.zeros(shape=[784 , 10]))
b = tf.Variable(tf.zeros(shape=[10]))
logits = tf.matmul(x,W) + b
y_pred = tf.nn.softmax(logits)

loss = tf.reduce_mean(-tf.reduce_sum(y * tf.log(y_pred), reduction_indices=[1]))
#loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=y)) # tf.nn.softmax_cross_entropy_with_logits API를 이용한 구현
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(loss)


sess = tf.Session()
sess.run(tf.global_variables_initializer())

for i in range(1000):
  batch_xs , batch_ys =mnist.train.next_batch(100)
  sess.run(train_step ,feed_dict = {x: batch_xs , y: batch_ys})
  correct_prediction = tf.equal(tf.argmax(y_pred,1), tf.argmax(y,1))
  accuracy =tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
  print("정확도:%f" % sess.run(accuracy, feed_dict={x: mnist.test.images ,y:mnist.test.labels}))
sess.close()
```
